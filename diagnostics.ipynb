```python
import pandas as pd
import numpy as np
```

```python
try:
  df = pd.read_csv("./mumbai_court_master_data.csv")
except FileNotFoundError as e:
  print(e)
```

```python
rows = len(df)
unique_cnrs = df['cnr_number'].nunique()
avg_hearings = rows / unique_cnrs if unique_cnrs > 0 else 0
rows, unique_cnrs, avg_hearings
```

```python
junk_values = ['Unknown', 'nan', 'ERROR', 'CASE_ID', 'TIMEOUT']
for col in df.columns:
  null_count = df[col].isnull().sum()
  junk_count = df[col].astype(str).apply(lambda x: any(j in x for j in junk_values)).sum()
  total_bad = null_count + junk_count
  percentage = (total_bad / rows) * 100
  print(f"   * {col:18}: {total_bad:,} issues ({percentage:.1f}%)")
```
